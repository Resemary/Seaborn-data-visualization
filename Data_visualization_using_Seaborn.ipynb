Importing the libraries


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd




```

Seaborn has some built-in dataset

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme(style='darkgrid')
#visualizing the data
sns.relplot(data=tips, x="total_bill", y="tip", col='time', hue='smoker' ,style='smoker',size='size' );
#hue=smoker
#Purpose: This argument tells Seaborn to color-code the points in the plot  based on the 'smoker' column in your dataset.
#style='smoker'
#Purpose: This argument tells Seaborn to use different marker styles for the points based on the 'smoker' column
#size='size'
#Purpose: This argument tells Seaborn to vary the size of the points based on the 'size' column in your dataset.
```

Distribution of the total bill

```python
sns.displot(data=tips, x='total_bill', kde= True);
```

```python
#loading the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

Heat Map

```python
#constructing a heat map
plt.figure(figsize=(10,10))
sns.pairplot(iris, hue='species');
```

Scatter Plot

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_width', hue='species', style='species');
```
